```python
import os
import shutil
import random

TRAINING_FILES_PATH = "/content/drive/MyDrive/SET_project_1/SER/data/RAV_TESS"
CREMA_ORIGINAL_FOLDER_PATH = "/content/drive/MyDrive/SET_project_1/SER/data/CREMA-D"

class CREMAPipeline:

    @staticmethod
    def create_CREMA_folders(path):
        counter = 0
        # 01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised

        label_conversion = {'01': 'N',
                            '03': 'H',
                            '04': 'S',
                            '05': 'A',
                            '06': 'F',
                            '07': 'D',}

        for subdir, dirs, files in os.walk(path):
            for filename in files:
                destination_path = TRAINING_FILES_PATH + '/Actor_29//'
                old_file_path = os.path.join(os.path.abspath(subdir), filename)

                # Separate base from extension
                base, extension = os.path.splitext(filename)

                for key, value in label_conversion.items():
                    if base[-6] == value:
                      print(f"{base[-6]} = {key}")
                      random_list = random.sample(range(10, 99), 7)
                      file_name = '-'.join([str(i) for i in random_list])
                      file_name_with_correct_emotion = file_name[:6] + key + file_name[8:] + extension
                      new_file_path = destination_path + file_name_with_correct_emotion
                      shutil.copy(old_file_path, new_file_path)
                      counter +=1

        print(f"Finished, file count: {counter}")

if __name__ == '__main__':
    CREMAPipeline.create_CREMA_folders(CREMA_ORIGINAL_FOLDER_PATH)
```
